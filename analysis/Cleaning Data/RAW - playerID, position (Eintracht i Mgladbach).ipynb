```python
import numpy as np
import seaborn as sns
import pandas as pd
from pandas import Series,DataFrame
import os

#Dohvacanje CSV datoteke za utakmice 
fileDirU = os.path.dirname(os.path.realpath('mozgalo_gladbach_eintracht_utakmice.csv'))
filenameU = os.path.join(fileDirU, '../../data/clean/mozgalo_gladbach_eintracht_utakmice.csv')
fileU = os.path.abspath(os.path.realpath(filenameU))

#Dohvacanje CSV datoteke za akcije
fileDirA = os.path.dirname(os.path.realpath('mozgalo_gladbach_eintracht_akcije.csv'))
filenameA = os.path.join(fileDirA, '../../data/clean/mozgalo_gladbach_eintracht_akcije.csv')
fileA = os.path.abspath(os.path.realpath(filenameA))

#Dohvacanje XLSX datoteke za akcije
fileDirC = os.path.dirname(os.path.realpath('Once - Lista akcija.xlsx'))
filenameC = os.path.join(fileDirA, '../../data/clean/Once - Lista akcija.xlsx')
fileC = os.path.abspath(os.path.realpath(filenameC))

#Dohvacanje XLSX datoteke za pozicije igrača
fileDirD = os.path.dirname(os.path.realpath('once_player_position.xlsx'))
filenameD = os.path.join(fileDirD, '../../data/clean/once_player_position.xlsx')
fileD = os.path.abspath(os.path.realpath(filenameD))
```

```python
dframe_akcije = pd.read_csv(fileA,sep=",")
dframe_utakmice = pd.read_csv(fileU,sep=",")
dframe_pozicije = pd.read_excel(fileD)
```

```python
df_allPlayerIDs = dframe_akcije[["gameID","playerID","playerPositionID","playerTeam"]]
#Dohvacanje podataka o svim igračima
df_allPlayerIDs
```

```python
#Brisanje nepotrebnih podataka iz tablice utakmice
dframe_utakmice = dframe_utakmice.drop(dframe_utakmice.columns[[range(3,21)]],axis =1)
#Spajanje podataka da bi za naše igrače dobili
df_allPlayerIDs = pd.merge(df_allPlayerIDs, dframe_utakmice,left_on="gameID",right_on="ID")
df_allPlayerIDs.drop("ID",axis=1,inplace=True)
```

```python
df_allPlayerIDs.head(10)
```

```python
df_allPlayerIDs = pd.merge(df_allPlayerIDs,dframe_pozicije,left_on="playerPositionID",right_on="ID")
df_allPlayerIDs.drop("ID",axis=1,inplace=True)
df_allPlayerIDs.drop("playerPositionID",axis=1,inplace=True)
df_allPlayerIDs.drop_duplicates(inplace=True)
```

```python
df_allPlayerIDs.sort_values("gameID",inplace=True)
df_allPlayerIDs.reset_index(inplace=True,drop=True)
df_allPlayerIDs
```

```python
df_allPlayerIDs.to_excel("allPlayerIDs.xls")
```

```python
#Formacije i igrači za momčad Eintracht Frankfurt i Borussia Mgladbach,
eintrachtFrankfurt = pd.DataFrame(columns=["gameID","playerID","playerTeam","domaci","gosti", "position"])
borussiaMgladbach = pd.DataFrame(columns=["gameID","playerID","playerTeam","domaci","gosti", "position"])

for index, item in df_allPlayerIDs.iterrows():
    if (item["playerTeam"] == "home"):
        if (item["domaci"] == "Eintracht Frankfurt"):
            eintrachtFrankfurt.loc[len(eintrachtFrankfurt)] = item
        if (item["domaci"] == "Borussia Mgladbach"):
            borussiaMgladbach.loc[len(borussiaMgladbach)] = item
            
    if (item["playerTeam"] == "away"):
        if (item["gosti"] == "Eintracht Frankfurt"):
            eintrachtFrankfurt.loc[len(eintrachtFrankfurt)] = item
        if (item["gosti"] == "Borussia Mgladbach"):
            borussiaMgladbach.loc[len(borussiaMgladbach)] = item
```

```python
eintrachtFrankfurt = eintrachtFrankfurt.drop_duplicates()
eintrachtFrankfurt.playerID = eintrachtFrankfurt.playerID.astype(int)
```
