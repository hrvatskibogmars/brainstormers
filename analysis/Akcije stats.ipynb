```python
#Uƒçitavanje popisa akcija dobivenih od mozgala
listaAkcija = pd.read_excel(fileC)
listaAkcija
```

```python
#Spajanje na≈°e liste akcije s nazivom akcije
akcijeUnique = pd.merge(actionStats_df,listaAkcija, left_on="actionID",right_on="ID")
```

```python
#Makivanje nepotrebnog stupca koji se pojavio spajanjem
akcijeUnique.drop("actionID",axis=1,inplace=True)
#Spremanje liste u excel
akcijeUnique.to_excel("akcije.xls")
```

```python
akcije = pd.read_excel("akcije.xls")
# Pogledati match statistics dokument
match_df = pd.read_csv("match_stats.csv",sep=",")
akcije["awayAkcije"] = 0
akcije["homeAkcije"] = 0
akcije["totalAkcije"] = 0
akcije
```

```python
for index, item in akcije.iterrows():
    for indexA, itemA in match_df.iterrows():
        if (item["ID"] == itemA["actionID"]):
            akcije.ix[index,"totalAkcije"] = akcije.ix[index,"totalAkcije"] + itemA["sumAction"] 
            akcije.ix[index,"homeAkcije"] = akcije.ix[index,"homeAkcije"] + itemA['home']
            akcije.ix[index,"awayAkcije"] = akcije.ix[index,"awayAkcije"] + itemA['away']
```

```python
akcije.sort_values(["totalAkcije"], inplace=True, ascending=False)
akcije
```
