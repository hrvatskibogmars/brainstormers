### Priprema podataka za SNA - za sve akcije Eintrachta

Provjera u kojem rasponu se kreću vrijednosti za atribut distance

```python
import numpy as np
import pandas as pd
import matplotlib as plt
import warnings
warnings.filterwarnings('ignore')

dfSve = pd.read_csv("mozgalo_gladbach_eintracht_akcije_spojeno.csv", sep = ',')

maxDistance = dfSve['distance'].max()
distance = dfSve['distance']
distanceArray = np.array(distance)
medianDistance = np.median(distanceArray)
minDistance = dfSve['distance'].min()

print("Max: %s" % maxDistance)
print("Median: %s" % medianDistance)
print("Min: %s" % minDistance)
```

Kodiranje (diskretizacija) atributa distance tako da gephi može prepoznati težine bridova

```python
import numpy as np
import pandas as pd
import matplotlib as plt
import warnings
warnings.filterwarnings('ignore')

for index, item in dfSve.iterrows():
    if(item['distance'] >= 0 and item['distance'] <= 11.17):
        w = 1
        dfSve.set_value(index,'weight',w)
    elif(item['distance'] >= 11.17 and item['distance'] <= 22.34):
        w = 2
        dfSve.set_value(index,'weight',w)
    elif(item['distance'] >= 22.34 and item['distance'] <= 33.51):
        w = 3
        dfSve.set_value(index,'weight',w)
    elif(item['distance'] >= 33.51 and item['distance'] <= 44.68):
        w = 4
        dfSve.set_value(index,'weight',w)
    elif(item['distance'] >= 44.68 and item['distance'] <= 55.85):
        w = 5
        dfSve.set_value(index,'weight',w)
    elif(item['distance'] >= 55.85 and item['distance'] <= 67.02):
        w = 6
        dfSve.set_value(index,'weight',w)
    elif(item['distance'] >= 67.02 and item['distance'] <= 78.19):
        w = 7
        dfSve.set_value(index,'weight',w)
    elif(item['distance'] >= 78.19 and item['distance'] <= 89.36):
        w = 8
        dfSve.set_value(index,'weight',w)
```

Filtriranje stupaca i akcija samo naših igrača

```python
import numpy as np
import pandas as pd
import matplotlib as plt
import warnings
warnings.filterwarnings('ignore')

igraci = [8570, 196633, 5335, 21022, 16520, 199745, 77888, 39442, 1802, 1790, 1770, 22870, 40700, 196661, 212914, 203576, 
          206101, 67519, 143853, 57720, 12956, 44079, 7466, 205986, 102895, 201624, 214449, 206102]

dfEintrachtAkcije = dfSve[dfSve['playerID'].isin(igraci)]
dfEintrachtAkcijeClean = dfEintrachtAkcije.drop(['gameID', 'actionID',
                                            'bindedActionID', 'actionOptionID',
                                            'playerTeam', 'goalZone', 'refPositionX',
                                            'refPositionY', 'startPossTime', 'endPossTime', 'gameTimeSection',
                                            'playerPositionID', 'targetPositionID', 'awayTeamSide', 'action_duration',
                                            'domaci', 'gosti', 'domaciFormacija', 'gostiFormacija',
                                            'gameDate', 'gameTime', 'stadium', 'referee', 'stadium', 'climate',
                                            'luna', 'temperature', 'viewers', 'domaciID', 'gostiID', 'homeTeamFormacijaID',
                                            'awayTeamFormacijaID', 'Pogodaka_domaci', 'Score_domaci', 'Pogodaka_gosti',
                                            'Score_gosti', 'actionStartTime', 'actionEndTime', 'touchNumber',
                                            'actionPosX1', 'actionPosX2', 'actionPosY1', 'actionPosY2',
                                            'ref_sector'], axis = 1)

dfEintrachtAkcijeClean1 = dfEintrachtAkcijeClean[dfEintrachtAkcijeClean['targetPlayerID'] != 0]
dfEintrachtAkcijeClean2 = dfEintrachtAkcijeClean1[(dfEintrachtAkcijeClean1['playerID'].isin(igraci)) & 
                                                  (dfEintrachtAkcijeClean1['targetPlayerID'].isin(igraci))]
dfEintrachtAkcijeClean3 = dfEintrachtAkcijeClean2[dfEintrachtAkcijeClean2['playerID'] != 
                                                  dfEintrachtAkcijeClean2['targetPlayerID']]
```

```python
dfEP = pd.read_csv("eintrachtP.csv", sep=',')

dfEintrachtIgraci = pd.merge(dfEintrachtAkcijeClean3, dfEP, left_on='playerID', right_on='playerID', how='left')
dfEintrachtIgraci2 = pd.merge(dfEintrachtIgraci, dfEP, left_on='targetPlayerID', right_on='playerID', how='left')
dfEintrachtIgraci3 = dfEintrachtIgraci2.drop('playerID_y', axis=1)
dfEintrachtIgraci4 = dfEintrachtIgraci3.rename(columns={'name_x':'playerName', 'name_y':'targetName', 'playerID_x':'playerID'})
dfEintrachtIgraci5 = dfEintrachtIgraci4.drop(['distance', 'ID', 'weight'], axis=1)
dfEintrachtIgraci5['weight'] = dfEintrachtIgraci5.groupby('playerID')['targetPlayerID'].transform(pd.Series.value_counts)
dfEintrachtIgraci5 = dfEintrachtIgraci5.drop_duplicates()
dfEintrachtIgraci5.to_csv("Eintracht_sna_ukupno.csv", sep=',', index=False)
dfEintrachtIgraci5
```
