```python
import numpy as np
import seaborn as sns
import pandas as pd
from pandas import Series,DataFrame
import os

#Dohvacanje CSV datoteke za akcije BM
fileDirA = os.path.dirname(os.path.realpath('borussiaMgladbachSveAkcije.xls'))
filenameA = os.path.join(fileDirA, '../../analysis/Hrvoje/BM/borussiaMgladbachSveAkcije.xls')
fileA = os.path.abspath(os.path.realpath(filenameA))


#Dohvacanje CSV datoteke za igrača BM
fileDirB = os.path.dirname(os.path.realpath('bmPlayers.xls'))
filenameB = os.path.join(fileDirB, '../../analysis/Hrvoje/BM/bmPlayers.xls')
fileB = os.path.abspath(os.path.realpath(filenameB))

#Dohvacanje CSV datoteke za igrača BM
fileDirC = os.path.dirname(os.path.realpath('eintrachtFrankfurtSveAkcije.xls'))
filenameC = os.path.join(fileDirC, '../../analysis/Hrvoje/EF/eintrachtFrankfurtSveAkcije.xls')
fileC = os.path.abspath(os.path.realpath(filenameC))

#Dohvacanje CSV datoteke za igrača BM
fileDirD = os.path.dirname(os.path.realpath('efPlayers.xls'))
filenameD = os.path.join(fileDirD, '../../analysis/Hrvoje/EF/efPlayers.xls')
fileD = os.path.abspath(os.path.realpath(filenameD))
```

```python
dframe_BM = pd.read_excel(fileA)
dframe_bmPlayers = pd.read_excel(fileB)

dframe_EF = pd.read_excel(fileC)
dframe_efPlayers = pd.read_excel(fileD)
```

```python
for indexA, itemA in dframe_bmPlayers.iterrows():
    goals = 0
    assists = 0
    for indexB,itemB in dframe_BM.iterrows():
        if (itemA["playerID"] == itemB["playerID"]):            
            goals += itemB["1"]
            assists += itemB["342"]
            
    dframe_bmPlayers.ix[indexA,"goals"] = goals
    dframe_bmPlayers.ix[indexA,"assists"] = assists
    
for indexA, itemA in dframe_efPlayers.iterrows():
    goals = 0
    assists = 0
    for indexB,itemB in dframe_EF.iterrows():
        if (itemA["playerID"] == itemB["playerID"]):            
            goals += itemB["1"]
            assists += itemB["342"]
            
    dframe_efPlayers.ix[indexA,"goals"] = goals
    dframe_efPlayers.ix[indexA,"assists"] = assists
    #print (str(itemA.playerID) +" G:" +str(goals) +" A:" +str(assists))
```
