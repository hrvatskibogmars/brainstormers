# Analiza momčadi

Analiza prikazuje preformanse momčadi u natjecanjima

### Opis tablice
* ID momčadi
* Ime momčadi
* Ukupan broj bodova (score)
* Broj pobjeda/poraza/izjednačenih utakmica (W/L/D)
* Broj pobjeda s obzirom na home / away
* Ukupan broj utakmica (totalMatches)
* Ukupan broj primljenih pogodaka (goalsScored) 
* Ukupan broj pogodaka (goalsRecived)

```python
import numpy as np
import seaborn as sns
import pandas as pd
from pandas import Series,DataFrame
import os

#Dohvacanje CSV datoteke za utakmice 
fileDirU = os.path.dirname(os.path.realpath('mozgalo_gladbach_eintracht_utakmice.csv'))
filenameU = os.path.join(fileDirU, '../../data/clean/mozgalo_gladbach_eintracht_utakmice.csv')
fileU = os.path.abspath(os.path.realpath(filenameU))

#Dohvacanje CSV datoteke za akcije
fileDirA = os.path.dirname(os.path.realpath('mozgalo_gladbach_eintracht_akcije.csv'))
filenameA = os.path.join(fileDirA, '../../data/clean/mozgalo_gladbach_eintracht_akcije.csv')
fileA = os.path.abspath(os.path.realpath(filenameA))
```

```python
dframe_utakmice = pd.read_csv(fileU, sep=",")
dframe_akcije = pd.read_csv(fileA, sep=",")
```

```python
momcadiA = dframe_utakmice[['domaci','domaciID',]]
momcadiB = dframe_utakmice[['gosti','gostiID']]
```

```python
momcadiA= momcadiA.drop_duplicates()
mocadiA = momcadiA.reset_index(drop=True, inplace=True)
momcadiA.columns = ['name','ID']
momcadiA
```

```python
momcadiB = momcadiB.drop_duplicates()
momcadiB.reset_index(drop=True, inplace=True)
momcadiB.columns = ['name','ID']
momcadiB
```

```python
result = momcadiA.append(momcadiB)
result.drop_duplicates(inplace=True)
result.reset_index(drop=True, inplace=True)
result
```

```python
test = dframe_utakmice[['domaci','Score_domaci','gosti','Score_gosti', 'Pogodaka_domaci','Pogodaka_gosti',]]
test
```

```python
for index, item in result.iterrows():
    formH = ""
    form = ""
    formA = ""
    f = 0
    l = 0
    #home lose
    hl = 0
    #away lose
    al = 0
    #total wins
    w = 0
    #home 
    hw = 0
    aw = 0
    d = 0
    hd = 0
    ad = 0
    totalGoals = 0
    primljeni = 0
    for i, row in test.iterrows():
        if item["name"] == row['domaci']:
            if row['Score_domaci'] == 0:
                l = l + 1
                hl = hl + 1
                form += "l"
                formH += "l"
            if row['Score_domaci'] == 1:
                d = d + 1
                hd = hd + 1
                form += "d"
                formH += "d"

            if row['Score_domaci'] == 3:
                w = w + 1
                hw = hw + 1
                form += "w"
                formH += "w"

            totalGoals = totalGoals + row['Pogodaka_domaci']
            primljeni = primljeni + row['Pogodaka_gosti']
            f = f +row["Score_domaci"]
            #print (item["name"])
            #print (f)
        if item["name"] == row['gosti']:
            if row['Score_gosti'] == 0:
                l = l + 1
                al = al + 1
                form += "l"
                formA += "l"
            if row['Score_gosti'] == 1:
                d = d + 1
                ad = ad + 1
                form += "d"
                formA += "d"
            if row['Score_gosti'] == 3:
                w = w + 1
                aw = aw + 1
                form += "w"
                formA += "w"
            f = f+ row["Score_gosti"]
            totalGoals = totalGoals + row['Pogodaka_gosti']
            primljeni = primljeni + row['Pogodaka_domaci']
    #print (item["name"] + ": " + str(f))
    #print ("w=" + str(w) + ", l="+ str(l) + ", d=" + str(d))
    result.ix[index,"w"] = w
    result.ix[index,"l"] = l
    result.ix[index,"d"] = d

    #print ("hw=" + str(hw) + ", hl=" + str(hl) + ", hd=" + str(hd))
    result.ix[index,"hw"] = hw
    result.ix[index,"hl"] = hl
    result.ix[index,"hd"] = hd

    #print ("aw=" + str(aw) + ", al=" + str(al) + ", ad=" + str(ad))
    result.ix[index,"aw"] = aw
    result.ix[index,"al"] = al
    result.ix[index,"ad"] = ad
    
    #print ("Ukupan broj utakmica: " + str(w+l+d))
    result.ix[index,"totalMatches"] = w+l+d
    
    #print ("Ukupan broj golova: " + str(totalGoals))
    result.ix[index,"goalsScored"] = totalGoals
    
    #print ("Ukupan broj primljenih pogodaka: " + str(primljeni))
    result.ix[index,"goalsRecived"] = primljeni
    
    result.ix[index,"form"] = form
    result.ix[index,"formA"] = formA
    result.ix[index,"formH"] = formH
    

    #print ('\n')
```
