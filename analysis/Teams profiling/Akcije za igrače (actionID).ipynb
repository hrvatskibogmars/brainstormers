```python
import numpy as np
import seaborn as sns
import pandas as pd
from pandas import Series,DataFrame
import os

#Dohvacanje CSV datoteke za utakmice 
fileDirU = os.path.dirname(os.path.realpath('mozgalo_gladbach_eintracht_utakmice.csv'))
filenameU = os.path.join(fileDirU, '../data/clean/mozgalo_gladbach_eintracht_utakmice.csv')
fileU = os.path.abspath(os.path.realpath(filenameU))

#Dohvacanje CSV datoteke za akcije
fileDirA = os.path.dirname(os.path.realpath('mozgalo_gladbach_eintracht_akcije.csv'))
filenameA = os.path.join(fileDirA, '../data/clean/mozgalo_gladbach_eintracht_akcije.csv')
fileA = os.path.abspath(os.path.realpath(filenameA))

#Dohvacanje XLSX datoteke za akcije
fileDirC = os.path.dirname(os.path.realpath('Once - Lista akcija.xlsx'))
filenameC = os.path.join(fileDirA, '../data/clean/Once - Lista akcija.xlsx')
fileC = os.path.abspath(os.path.realpath(filenameC))

#Dohvacanje XLSX datoteke za pozicije igrača
fileDirD = os.path.dirname(os.path.realpath('once_player_position.xlsx'))
filenameD = os.path.join(fileDirD, '../data/clean/once_player_position.xlsx')
fileD = os.path.abspath(os.path.realpath(filenameD))
```

```python
dframe_eintrachtFrankfurt = pd.read_excel("eintrachtFrankfurtTekme.xls")
dframe_borussiaMgladbach = pd.read_excel("borussiaMgladbachTekme.xls")
dframe_akcije = pd.read_csv(fileA)
dframe_utakmice = pd.read_csv(fileU)
```

```python
#dframe_utakmice.head()#dframe_borussiaMgladbach.loc[dframe_borussiaMgladbach["gameID"] == 6418]
```

```python
for item in dframe_akcije["actionID"].unique():
    dframe_borussiaMgladbach[str(item)] = 0

for item in dframe_akcije["actionID"].unique():
    dframe_eintrachtFrankfurt[str(item)] = 0
#dframe_borussiaMgladbach
```

```python
for indexA,itemA in dframe_borussiaMgladbach.iterrows():
    for indexB, itemB in dframe_utakmice.iterrows():
        if (itemA["gameID"] == itemB["ID"]):
            if (itemB["domaci"] == "Borussia Mgladbach"):
                if (itemB["Score_domaci"] == 3):
                    dframe_borussiaMgladbach.loc[indexA,"matchStatus"] = "w"
                if (itemB["Score_domaci"] == 1):
                    dframe_borussiaMgladbach.loc[indexA,"matchStatus"] = "d"
                if (itemB["Score_domaci"] == 0):
                    dframe_borussiaMgladbach.loc[indexA,"matchStatus"] = "l"
            if (itemB["gosti"] == "Borussia Mgladbach"):
                if (itemB["Score_gosti"] == 3):
                    dframe_borussiaMgladbach.loc[indexA,"matchStatus"] = "w"
                if (itemB["Score_gosti"] == 1):
                    dframe_borussiaMgladbach.loc[indexA,"matchStatus"] = "d"
                if (itemB["Score_gosti"] == 0):
                    dframe_borussiaMgladbach.loc[indexA,"matchStatus"] = "l"
dframe_borussiaMgladbach
```

```python
for indexA,itemA in dframe_eintrachtFrankfurt.iterrows():
    for indexB, itemB in dframe_utakmice.iterrows():
        if (itemA["gameID"] == itemB["ID"]):
            if (itemB["domaci"] == "Eintracht Frankfurt"):
                if (itemB["Score_domaci"] == 3):
                    dframe_eintrachtFrankfurt.loc[indexA,"matchStatus"] = "w"
                if (itemB["Score_domaci"] == 1):
                    dframe_eintrachtFrankfurt.loc[indexA,"matchStatus"] = "d"
                if (itemB["Score_domaci"] == 0):
                    dframe_eintrachtFrankfurt.loc[indexA,"matchStatus"] = "l"
            if (itemB["gosti"] == "Eintracht Frankfurt"):
                if (itemB["Score_gosti"] == 3):
                    dframe_eintrachtFrankfurt.loc[indexA,"matchStatus"] = "w"
                if (itemB["Score_gosti"] == 1):
                    dframe_eintrachtFrankfurt.loc[indexA,"matchStatus"] = "d"
                if (itemB["Score_gosti"] == 0):
                    dframe_eintrachtFrankfurt.loc[indexA,"matchStatus"] = "l"
```

```python
#for indexA,item in dframe_borussiaMgladbach.iterrows():
for indexA,itemA in dframe_borussiaMgladbach.iterrows():
    for indexB, itemB in dframe_akcije.iterrows():
        if (itemA["gameID"] == itemB["gameID"] and itemA["playerID"] == itemB["playerID"]):
            dframe_borussiaMgladbach.loc[indexA,str(itemB.actionID)] +=1
            #print ("h")
```
