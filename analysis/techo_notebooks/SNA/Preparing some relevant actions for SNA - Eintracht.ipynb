Priprema podataka za SNA o kontra napadima

```python
import numpy as np
import pandas as pd
import matplotlib as plt
import warnings
warnings.filterwarnings('ignore')

dfSve = pd.read_csv("mozgalo_gladbach_eintracht_akcije_spojeno.csv", sep = ',')

igraci = [8570, 196633, 5335, 21022, 16520, 199745, 77888, 39442, 1802, 1790, 1770, 22870, 40700, 196661, 212914, 203576, 
          206101, 67519, 143853, 57720, 12956, 44079, 7466, 205986, 102895, 201624, 214449, 206102]

dfEintrachtAkcije = dfSve[dfSve['playerID'].isin(igraci)]
dfEintrachtKontraNapadi = dfEintrachtAkcije[(dfEintrachtAkcije['actionID'] == 12) & 
                                            (dfEintrachtAkcije['actionOptionID'] == 354)]

dfEintrachtKontraNapadiClean = dfEintrachtKontraNapadi.drop(['ID', 'gameID', 'actionID',
                                            'bindedActionID', 'actionOptionID',
                                            'playerTeam', 'goalZone', 'refPositionX',
                                            'refPositionY', 'startPossTime', 'endPossTime', 'gameTimeSection',
                                            'playerPositionID', 'targetPositionID', 'awayTeamSide', 'action_duration',
                                            'domaci', 'gosti', 'domaciFormacija', 'gostiFormacija',
                                            'gameDate', 'gameTime', 'stadium', 'referee', 'stadium', 'climate',
                                            'luna', 'temperature', 'viewers', 'domaciID', 'gostiID', 'homeTeamFormacijaID',
                                            'awayTeamFormacijaID', 'Pogodaka_domaci', 'Score_domaci', 'Pogodaka_gosti',
                                            'Score_gosti', 'actionStartTime', 'actionEndTime', 'touchNumber',
                                            'actionPosX1', 'actionPosX2', 'actionPosY1', 'actionPosY2',
                                            'ref_sector'], axis = 1)

dfEintrachtKontraNapadiClean.reset_index(drop=True, inplace=True)

dfEintrachtKontraNapadiClean1 = dfEintrachtKontraNapadiClean[dfEintrachtKontraNapadiClean['targetPlayerID'] != 0]
dfEintrachtKontraNapadiClean2 = dfEintrachtKontraNapadiClean1[(dfEintrachtKontraNapadiClean1['playerID'].isin(igraci)) & 
                                                  (dfEintrachtKontraNapadiClean1['targetPlayerID'].isin(igraci))]
dfEintrachtKontraNapadiClean3 = dfEintrachtKontraNapadiClean2[dfEintrachtKontraNapadiClean2['playerID'] != 
                                                  dfEintrachtKontraNapadiClean2['targetPlayerID']]

dfEP = pd.read_csv("eintrachtP.csv", sep=',')

dfEintrachtIgraci = pd.merge(dfEintrachtKontraNapadiClean3, dfEP, left_on='playerID', right_on='playerID', how='left')
dfEintrachtIgraci2 = pd.merge(dfEintrachtIgraci, dfEP, left_on='targetPlayerID', right_on='playerID', how='left')
dfEintrachtIgraci3 = dfEintrachtIgraci2.drop('playerID_y', axis=1)
dfEintrachtIgraci4 = dfEintrachtIgraci3.rename(columns={'name_x':'playerName', 'name_y':'targetName', 'playerID_x':'playerID'})
dfEintrachtIgraci5 = dfEintrachtIgraci4.drop(['distance'], axis=1)
dfEintrachtIgraci5['Weight'] = dfEintrachtIgraci5.groupby('playerID')['targetPlayerID'].transform(pd.Series.value_counts)
dfEintrachtIgraci5 = dfEintrachtIgraci5.drop_duplicates()
dfEintrachtIgraci5.to_csv("Eintracht_sna_kontra_napadi.csv", sep=',', index=False)
dfEintrachtIgraci5
```

Priprema ƒçvorova

```python
dfNodes = dfEintrachtIgraci5[['playerID', 'playerName']]
dfNodesUnique = dfNodes.drop_duplicates()
dfNodesUniqueRenamed = dfNodesUnique.rename(columns={'playerID':'ID', 'playerName':'Label'})
dfNodesUniqueRenamed.to_csv("Eintracht_sna__kontre_nodes.csv", sep=',', index=False)
```

Priprema bridova

```python
dfEdges = dfEintrachtIgraci5.rename(columns={'playerID':'Source', 'targetPlayerID':'Target', 'weight':'Weight'})
dfEdges.to_csv("Eintracht_sna_kontre_edges.csv", sep=',', index=False)
```

Priprema podataka o atraktivnim udarcima (u okvir i izvan okvira)

```python
dfEintrachtAtrUdarci = dfEintrachtAkcije[((dfEintrachtAkcije['actionID'] == 30) & 
                                            (dfEintrachtAkcije['bindedActionID'] == 31)) |
                                        ((dfEintrachtAkcije['actionID'] == 30) & 
                                            (dfEintrachtAkcije['bindedActionID'] == 262))]
dfEintrachtAtrUdarciClean = dfEintrachtAtrUdarci.drop(['ID','gameID', 'actionID',
                                            'bindedActionID', 'actionOptionID',
                                            'playerTeam', 'goalZone', 'refPositionX',
                                            'refPositionY', 'startPossTime', 'endPossTime', 'gameTimeSection',
                                            'playerPositionID', 'targetPositionID', 'awayTeamSide', 'action_duration',
                                            'domaci', 'gosti', 'domaciFormacija', 'gostiFormacija',
                                            'gameDate', 'gameTime', 'stadium', 'referee', 'stadium', 'climate',
                                            'luna', 'temperature', 'viewers', 'domaciID', 'gostiID', 'homeTeamFormacijaID',
                                            'awayTeamFormacijaID', 'Pogodaka_domaci', 'Score_domaci', 'Pogodaka_gosti',
                                            'Score_gosti', 'actionStartTime', 'actionEndTime', 'touchNumber',
                                            'actionPosX1', 'actionPosX2', 'actionPosY1', 'actionPosY2',
                                            'ref_sector'], axis = 1)

dfEintrachtAtrUdarciClean.reset_index(drop=True, inplace=True)

dfEP = pd.read_csv("eintrachtP.csv", sep=',')

dfEintrachtu = pd.merge(dfEintrachtAtrUdarciClean, dfEP, left_on='playerID', right_on='playerID', how='left')
dfEintrachtIgraciu1 = pd.merge(dfEintrachtu, dfEP, left_on='targetPlayerID', right_on='playerID', how='left')
dfEintrachtIgraciu2 = dfEintrachtIgraciu1.drop('playerID_y', axis=1)
dfEintrachtIgraciu3 = dfEintrachtIgraciu2.rename(columns={'name_x':'playerName', 'name_y':'targetName', 'playerID_x':'playerID'})
dfEintrachtIgraciu3.index += 1
dfEintrachtIgraciu3.to_csv("Eintracht_sna_udarci.csv", sep=',', index=False)
dfEintrachtIgraciu3
```
