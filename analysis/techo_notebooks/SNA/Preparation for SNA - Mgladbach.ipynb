```python
import numpy as np
import pandas as pd
import matplotlib as plt
import warnings
warnings.filterwarnings('ignore')

dfSve = pd.read_csv("mozgalo_gladbach_eintracht_akcije_spojeno.csv", sep = ',')

maxDistance = dfSve['distance'].max()
distance = dfSve['distance']
distanceArray = np.array(distance)
medianDistance = np.median(distanceArray)
minDistance = dfSve['distance'].min()

print("Max: %s" % maxDistance)
print("Median: %s" % medianDistance)
print("Min: %s" % minDistance)
```

```python
import numpy as np
import pandas as pd
import matplotlib as plt
import warnings
warnings.filterwarnings('ignore')

for index, item in dfSve.iterrows():
    if(item['distance'] >= 0 and item['distance'] <= 11.17):
        w = 1
        dfSve.set_value(index,'weight',w)
    elif(item['distance'] >= 11.17 and item['distance'] <= 22.34):
        w = 2
        dfSve.set_value(index,'weight',w)
    elif(item['distance'] >= 22.34 and item['distance'] <= 33.51):
        w = 3
        dfSve.set_value(index,'weight',w)
    elif(item['distance'] >= 33.51 and item['distance'] <= 44.68):
        w = 4
        dfSve.set_value(index,'weight',w)
    elif(item['distance'] >= 44.68 and item['distance'] <= 55.85):
        w = 5
        dfSve.set_value(index,'weight',w)
    elif(item['distance'] >= 55.85 and item['distance'] <= 67.02):
        w = 6
        dfSve.set_value(index,'weight',w)
    elif(item['distance'] >= 67.02 and item['distance'] <= 78.19):
        w = 7
        dfSve.set_value(index,'weight',w)
    elif(item['distance'] >= 78.19 and item['distance'] <= 89.36):
        w = 8
        dfSve.set_value(index,'weight',w)
```

Filtriranje stupaca i akcija samo naših igrača

```python
import numpy as np
import pandas as pd
import matplotlib as plt
import warnings
warnings.filterwarnings('ignore')

igraci = [16087, 206309, 11341, 11500, 16894, 4610, 39479, 205485, 205180, 102647, 156709, 115078, 126631, 16491, 127794, 185121,
          175582, 206094, 101192, 16901, 65256, 4613, 66692]

dfMgladbachAkcije = dfSve[dfSve['playerID'].isin(igraci)]
dfMgladbachAkcijeClean = dfMgladbachAkcije.drop(['gameID', 'actionID',
                                            'bindedActionID', 'actionOptionID',
                                            'playerTeam', 'goalZone', 'refPositionX',
                                            'refPositionY', 'startPossTime', 'endPossTime', 'gameTimeSection',
                                            'playerPositionID', 'targetPositionID', 'awayTeamSide', 'action_duration',
                                            'domaci', 'gosti', 'domaciFormacija', 'gostiFormacija',
                                            'gameDate', 'gameTime', 'stadium', 'referee', 'stadium', 'climate',
                                            'luna', 'temperature', 'viewers', 'domaciID', 'gostiID', 'homeTeamFormacijaID',
                                            'awayTeamFormacijaID', 'Pogodaka_domaci', 'Score_domaci', 'Pogodaka_gosti',
                                            'Score_gosti', 'actionStartTime', 'actionEndTime', 'touchNumber',
                                            'actionPosX1', 'actionPosX2', 'actionPosY1', 'actionPosY2',
                                            'ref_sector'], axis = 1)

dfMgladbachAkcijeClean1 = dfMgladbachAkcijeClean[dfMgladbachAkcijeClean['targetPlayerID'] != 0]
dfMgladbachAkcijeClean2 = dfMgladbachAkcijeClean1[(dfMgladbachAkcijeClean1['playerID'].isin(igraci)) & 
                                                  (dfMgladbachAkcijeClean1['targetPlayerID'].isin(igraci))]
dfMgladbachAkcijeClean3 = dfMgladbachAkcijeClean2[dfMgladbachAkcijeClean2['playerID'] != 
                                                  dfMgladbachAkcijeClean2['targetPlayerID']]
```

```python
dfEP = pd.read_csv("mgladbachP.csv", sep=',')

dfMgladbachIgraci = pd.merge(dfMgladbachAkcijeClean3, dfEP, left_on='playerID', right_on='playerID', how='left')
dfMgladbachIgraci2 = pd.merge(dfMgladbachIgraci, dfEP, left_on='targetPlayerID', right_on='playerID', how='left')
dfMgladbachIgraci3 = dfMgladbachIgraci2.drop('playerID_y', axis=1)
dfMgladbachIgraci4 = dfMgladbachIgraci3.rename(columns={'name_x':'playerName', 'name_y':'targetName', 'playerID_x':'playerID'})
dfMgladbachIgraci5 = dfMgladbachIgraci4.drop(['distance', 'ID', 'weight'], axis=1)
dfMgladbachIgraci5['weight'] = dfMgladbachIgraci5.groupby('playerID')['targetPlayerID'].transform(pd.Series.value_counts)
dfMgladbachIgraci5 = dfMgladbachIgraci5.drop_duplicates()
dfMgladbachIgraci5.to_csv("Mgladbach_sna_ukupno.csv", sep=',', index=False)
dfMgladbachIgraci5
```
