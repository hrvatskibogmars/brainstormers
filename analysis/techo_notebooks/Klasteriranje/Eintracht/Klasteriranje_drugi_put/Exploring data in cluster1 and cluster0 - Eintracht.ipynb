# Istraživanje klastera 1 sa akcijama koje su započele u sekotru 1 i 5

```python
import numpy as np
import pandas as pd
import matplotlib as plt
import warnings
warnings.filterwarnings('ignore')

dfKlaster1 = pd.read_csv("klaster1.csv", sep=',')
dfKlaster1
```

# Pobrojane akcije iz start sektora 1

```python
dfStartSektor1 = dfKlaster1[dfKlaster1['start_sector'] == 1]
dfStartSektor1.groupby(['actionIDNaziv']).size()
```

# Pobrojane akcije iz start sektora 5

```python
dfStartSektor5 = dfKlaster1[dfKlaster1['start_sector'] == 5]
dfStartSektor5.groupby('actionIDNaziv').size()
```

Akcije koje su nam zanimljive iz navedenih sektora su akcije CENTARŠUT, SKOK GLAVOM, DUEL IGRAČA, LOPTA PO ZRAKU. To su akcije Eintrachta koje mogu rezultirati pogotkom. Idemo vidjeti koji igrači sudjeluju u tim akcijama.

```python
dfRelAkcije = dfStartSektor1[['ID', 'actionIDNaziv', 'actionOptionIDN', 'playerID', 'actionPosX1', 'actionPosY1']]
dfRelAkcije1 = dfRelAkcije[(dfRelAkcije['actionIDNaziv'] == 'CENTARSUT')]
dfRelAkcije1
```

Spajanje dataframea sa podacima o imenima igrača

```python
dfEintrachtP = pd.read_csv("eintrachtP.csv", sep=',')
dfRelAkcije2 = pd.merge(dfRelAkcije1, dfEintrachtP, left_on='playerID', right_on='playerID', how='left')
dfRelAkcije2.to_csv("Eintracht_centarsutevi_sek1.csv", sep=',', index=False)
dfRelAkcije2
```
