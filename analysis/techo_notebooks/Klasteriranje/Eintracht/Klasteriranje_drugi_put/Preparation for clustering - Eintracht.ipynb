```python
import numpy as np
import seaborn as sns
import pandas as pd
from pandas import Series,DataFrame
import os

igraci = [8570, 196633, 5335, 21022, 16520, 199745, 77888, 39442, 1802, 1790, 1770, 22870, 40700, 196661, 212914, 203576, 
          206101, 67519, 143853, 57720, 12956, 44079, 7466, 205986, 102895, 201624, 214449, 206102]

dfAkcije = pd.read_csv("mozgalo_gladbach_eintracht_akcije_spojeno.csv", sep = ',')
dfUtakmice = pd.read_csv("mozgalo_gladbach_eintracht_utakmice.csv", sep = ',')
dfSpojeno = pd.merge(dfAkcije, dfUtakmice, left_on = 'gameID', right_on = 'ID')
dfClean = dfSpojeno.drop(['ID_y'], axis=1)
dfClean.rename(columns={'ID_x':'ID'}, inplace=True)
dfEintrachtUkupno = dfClean[dfClean['playerID'].isin(igraci)]
dfEintrachtUkupno.to_csv("Eintracht_ukupno2.csv", sep=',', index=False)
dfEintrachtUkupno
```
