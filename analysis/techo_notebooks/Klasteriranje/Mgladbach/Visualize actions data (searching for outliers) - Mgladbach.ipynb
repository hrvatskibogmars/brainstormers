```python
import matplotlib
import pandas as pd
import matplotlib.pyplot as plt
from pandas.tools.plotting import scatter_matrix
import warnings
warnings.filterwarnings('ignore')

dfMgladbachUkupno = pd.read_csv("Mgladbach_za_klasteriranje.csv", sep =',')

dfMgladbachClean =   dfMgladbachUkupno.drop(['ID', 'gameID', 'actionID',
                                            'bindedActionID', 'actionOptionID', 'playerID', 'actionIDNaziv',
                                            'bindedActionIDN', 'actionOptionIDN', 'targetPlayerID', 'playerPositionIDN',
                                            'targetPositionIDN', 'playerTeam', 'goalZone', 'refPositionX',
                                            'refPositionY', 'startPossTime', 'endPossTime', 'gameTimeSection',
                                            'playerPositionID', 'targetPositionID', 'awayTeamSide', 'action_duration',
                                            'poss_duration', 'domaci', 'gosti', 'domaciFormacija', 'gostiFormacija',
                                            'gameDate', 'gameTime', 'stadium', 'referee', 'stadium', 'climate',
                                            'luna', 'temperature', 'viewers', 'domaciID', 'gostiID', 'homeTeamFormacijaID',
                                            'awayTeamFormacijaID', 'Pogodaka_domaci', 'Score_domaci', 'Pogodaka_gosti',
                                            'Score_gosti'], axis = 1)

dfMgladbachClean
```

```python
fig1 = plt.figure(1, figsize = (22,22))
ax = fig1.gca()
sm = scatter_matrix(dfMgladbachClean, alpha=0.2, figsize=(15, 15), diagonal='hist', ax=ax)

[s.xaxis.label.set_rotation(45) for s in sm.reshape(-1)]
[s.yaxis.label.set_rotation(0) for s in sm.reshape(-1)]

[s.get_yaxis().set_label_coords(-0.7,0.5) for s in sm.reshape(-1)]
fig1.savefig('ScatterMatrica.png')
plt.show()
```
