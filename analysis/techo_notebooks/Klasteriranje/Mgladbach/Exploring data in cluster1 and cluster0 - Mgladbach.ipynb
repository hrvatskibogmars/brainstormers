# Istraživanje klastera 1 sa akcijama koje su započele u sekotru 1 i 5

```python
import numpy as np
import pandas as pd
import matplotlib as plt
import warnings
warnings.filterwarnings('ignore')

dfKlaster1 = pd.read_csv("klaster1_Mgladbach.csv", sep=',')
dfKlaster1
```

# Pobrojane akcije iz start sektora 13 i 16

```python
dfStartSektor13i16 = dfKlaster1[(dfKlaster1['start_sector'] == 13) | (dfKlaster1['start_sector'] == 16)]
dfStartSektor13i16.groupby(['actionIDNaziv']).size()
```

Akcije koje su nam zanimljive iz navedenih sektora su akcije CENTARŠUT, LOPTA PO ZRAKU. To su akcije Mgladbacha koje mogu rezultirati pogotkom. Idemo vidjeti koji igrači sudjeluju u tim akcijama.

```python
dfRelAkcije = dfStartSektor13i16[['ID', 'actionIDNaziv', 'actionOptionIDN', 'playerID', 'actionPosX1', 'actionPosY1', 'actionPosX2',
                                 'actionPosY2']]
dfRelAkcije1 = dfRelAkcije[(dfRelAkcije['actionIDNaziv'] == 'CENTARSUT')]
dfRelAkcije1
```

Spajanje dataframea sa podacima o imenima igrača

```python
dfMgladbachP = pd.read_csv("mgladbachP.csv", sep=',')
dfRelAkcije2 = pd.merge(dfRelAkcije1, dfMgladbachP, left_on='playerID', right_on='playerID', how='left')
dfRelAkcije2.to_csv("Mgladbach_centarsutevi_sek13i16.csv", sep=',', index=False)
dfRelAkcije2
```

# Pregled podataka iz klastera 0 u kojem se nalazi najviše akcija koje su počele u sektoru 1 i 5

```python
dfKlaster0 = pd.read_csv("klaster0_Mgladbach.csv", sep=',')
dfKlaster0
```

Ovjde nas isto zanimaju akcije koje su počele u sektoru 5 i koje bi mogle rezultirati pogotkom, a tu su udarci koji se ističu po broju svih akcija.

```python
dfStartSektor1i5 = dfKlaster0[(dfKlaster0['start_sector'] == 1) | (dfKlaster0['start_sector'] == 5)]
dfStartSektor1i5.groupby(['actionIDNaziv']).size()
```

```python
dfRelAkcije1 = dfStartSektor1i5[['ID', 'actionIDNaziv', 'actionOptionIDN', 'playerID', 'actionPosX1', 'actionPosY1', 'actionPosX2',
                                'actionPosY2']]
dfRelAkcije3 = dfRelAkcije1[(dfRelAkcije1['actionIDNaziv'] == 'UDARAC')]
dfRelAkcije3
```

```python
dfEintrachtP = pd.read_csv("mgladbachP.csv", sep=',')
dfRelAkcije4 = pd.merge(dfRelAkcije3, dfEintrachtP, left_on='playerID', right_on='playerID', how='left')
dfRelAkcije4.to_csv("Mgladbach_udarci_sek5.csv", sep=',', index=False)
dfRelAkcije4
```
