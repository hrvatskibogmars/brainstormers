# Vizualizacija bitnih akcija iz klastera 0 i 1

Učitavanje .csv filea u dataframe

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')

dfSek13i16 = pd.read_csv("Mgladbach_centarsutevi_sek13i16.csv", sep = ',')
dfSek13i16
```

Vizualizacija početka i kraja centaršuteva iz sektora 13 i 16

```python
dfx1 = dfSek13i16['actionPosX1']
dfy1 = dfSek13i16['actionPosY1']
dfx2 = dfSek13i16['actionPosX2']
dfy2 = dfSek13i16['actionPosY2']

fig = plt.figure(figsize = (12, 24))
fig.clf()
ax = fig.add_subplot(1,1,1)

plt.plot([dfx1, dfx2], [dfy1, dfy2], c='black')
plt.yticks([0,225,450,675,900])
plt.xticks([0,350,700,1050,1400])
im = plt.imread('igraliste2.png')
ax.scatter(dfx1, dfy1, s=60, zorder=1, c='b', label='Početak')
ax.scatter(dfx2, dfy2, s=60, zorder=1, c='y', label='Kraj')
implot = plt.imshow(im, zorder=0, extent=[0, 1400, 900, 0])
plt.xlabel('actionPosX1')
plt.ylabel('actionPosY1')
plt.title('Lokacije početka i kraja centaršutova - Mgladbach sektor 13 i 16')
plt.legend()
plt.show()
```

# Vizualizacija početka i kraja svih centaršuteva

```python
dfSveM = pd.read_csv("Mgladbach_sve_akcije.csv", sep = ',')
dfSveM
```

Vizualizacija svih centaršuteva za Mgladbach

```python
dfCentarsutevi = dfSveM[dfSveM['actionIDNaziv'] == 'CENTARSUT']
dfCentarsuteviFiltered = dfCentarsutevi[['ID', 'actionIDNaziv', 'actionOptionIDN', 'playerID', 'actionPosX1', 'actionPosY1', 
                                         'actionPosX2', 'actionPosY2']]

dfx1 = dfCentarsuteviFiltered['actionPosX1']
dfy1 = dfCentarsuteviFiltered['actionPosY1']
dfx2 = dfCentarsuteviFiltered['actionPosX2']
dfy2 = dfCentarsuteviFiltered['actionPosY2']

fig = plt.figure(figsize = (12, 24))
fig.clf()
ax = fig.add_subplot(1,1,1)

plt.plot([dfx1, dfx2], [dfy1, dfy2], c='black')
plt.yticks([0,225,450,675,900])
plt.xticks([0,350,700,1050,1400])
im = plt.imread('igraliste2.png')
ax.scatter(dfx1, dfy1, s=60, zorder=1, c='b', label='Početak')
ax.scatter(dfx2, dfy2, s=60, zorder=1, c='y', label='Kraj')
implot = plt.imshow(im, zorder=0, extent=[0, 1400, 900, 0])
plt.xlabel('actionPosX1')
plt.ylabel('actionPosY1')
plt.title('Lokacije početka i kraja  svih centaršutova - Mgladbach')
plt.legend(loc='lower left')
plt.show()
```
