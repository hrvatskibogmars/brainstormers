# Vizualizacija bitnih akcija iz klastera 0 i 1

Uƒçitavanje .csv filea u dataframe

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')

dfSek13i16 = pd.read_csv("Mgladbach_centarsutevi_sek13i16.csv", sep = ',')
dfSek13i16
```

Vizualizacija poƒçetka i kraja centar≈°uteva iz sektora 13 i 16

```python
dfDoma = dfSek13i16[dfSek13i16['domaci'] == 'Borussia Mgladbach']
dfGosti = dfSek13i16[dfSek13i16['gosti'] == 'Borussia Mgladbach']

dfx1Doma = dfDoma['actionPosX1']
dfy1Doma = dfDoma['actionPosY1']
dfx2Doma = dfDoma['actionPosX2']
dfy2Doma = dfDoma['actionPosY2']

dfx1Gosti = dfGosti['actionPosX1']
dfy1Gosti = dfGosti['actionPosY1']
dfx2Gosti = dfGosti['actionPosX2']
dfy2Gosti = dfGosti['actionPosY2']

fig = plt.figure(figsize = (12, 24))
fig.clf()
ax1 = fig.add_subplot(1,1,1)
ax2 = fig.add_subplot(1,1,1)
ax3 = fig.add_subplot(1,1,1)
ax4 = fig.add_subplot(1,1,1)

plt.plot([dfx1Doma, dfx2Doma], [dfy1Doma, dfy2Doma], c='black')
plt.plot([dfx1Gosti, dfx2Gosti], [dfy1Gosti, dfy2Gosti], c='black')
plt.yticks([0,225,450,675,900])
plt.xticks([0,350,700,1050,1400])
im = plt.imread('igraliste2.png')
ax1.scatter(dfx1Doma, dfy1Doma, s=60, zorder=1, c='b', label='Poƒçetak doma')
ax2.scatter(dfx2Doma, dfy2Doma, s=60, zorder=1, c='y', label='Kraj doma')
ax3.scatter(dfx1Gosti, dfy1Gosti, s=60, zorder=1, c='r', label='Poƒçetak gosti')
ax4.scatter(dfx2Gosti, dfy2Gosti, s=60, zorder=1, c='c', label='Kraj gosti')

implot = plt.imshow(im, zorder=0, extent=[0, 1400, 900, 0])
plt.xlabel('actionPosX')
plt.ylabel('actionPosY')
plt.title('Lokacije poƒçetka i kraja centar≈°utova prema koordinatama - Mgladbach sektor 13 i 16')

def label_point(x, y, val, ax):
    a = pd.concat({'x': x, 'y': y, 'val': val}, axis=1)
    for i, point in a.iterrows():
        ax.text(point['x'], point['y'], str(point['val']))

label_point(dfSek13i16.actionPosX1, dfSek13i16.actionPosY1, dfSek13i16.name, ax1)
label_point(dfSek13i16.actionPosX1, dfSek13i16.actionPosY1, dfSek13i16.name, ax1)

plt.legend()
plt.show()
```

# Vizualizacija poƒçetka i kraja svih centar≈°uteva

```python
dfSveM = pd.read_csv("Mgladbach_sve_akcije.csv", sep = ',')
dfSveM
```

Vizualizacija svih centar≈°uteva za Mgladbach

```python
dfCentarsutevi = dfSveM[dfSveM['actionIDNaziv'] == 'CENTARSUT']
dfCentarsuteviFiltered = dfCentarsutevi[['ID', 'actionIDNaziv', 'actionOptionIDN', 'playerID', 'actionPosX1', 'actionPosY1', 
                                         'actionPosX2', 'actionPosY2', 'domaci', 'gosti']]

dfDoma = dfCentarsuteviFiltered[dfCentarsuteviFiltered['domaci'] == 'Borussia Mgladbach']
dfGosti = dfCentarsuteviFiltered[dfCentarsuteviFiltered['gosti'] == 'Borussia Mgladbach']

dfx1Doma = dfDoma['actionPosX1']
dfy1Doma = dfDoma['actionPosY1']
dfx2Doma = dfDoma['actionPosX2']
dfy2Doma = dfDoma['actionPosY2']

dfx1Gosti = dfGosti['actionPosX1']
dfy1Gosti = dfGosti['actionPosY1']
dfx2Gosti = dfGosti['actionPosX2']
dfy2Gosti = dfGosti['actionPosY2']

fig = plt.figure(figsize = (12, 24))
fig.clf()
ax1 = fig.add_subplot(1,1,1)
ax2 = fig.add_subplot(1,1,1)
ax3 = fig.add_subplot(1,1,1)
ax4 = fig.add_subplot(1,1,1)

plt.plot([dfx1Doma, dfx2Doma], [dfy1Doma, dfy2Doma], c='black')
plt.plot([dfx1Gosti, dfx2Gosti], [dfy1Gosti, dfy2Gosti], c='black')
plt.yticks([0,225,450,675,900])
plt.xticks([0,350,700,1050,1400])
im = plt.imread('igraliste2.png')
ax1.scatter(dfx1Doma, dfy1Doma, s=60, zorder=1, c='b', label='Poƒçetak doma')
ax2.scatter(dfx2Doma, dfy2Doma, s=60, zorder=1, c='y', label='Kraj doma')
ax3.scatter(dfx1Gosti, dfy1Gosti, s=60, zorder=1, c='r', label='Poƒçetak gosti')
ax4.scatter(dfx2Gosti, dfy2Gosti, s=60, zorder=1, c='c', label='Kraj gosti')

implot = plt.imshow(im, zorder=0, extent=[0, 1400, 900, 0])
plt.xlabel('actionPosX')
plt.ylabel('actionPosY')
plt.title('Lokacije poƒçetka i kraja centar≈°utova prema koordinatama - Mgladbach ukupno')

plt.legend(loc='upper center')
plt.show()
```
