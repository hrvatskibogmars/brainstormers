```python
import numpy as np
import seaborn as sns
import pandas as pd
from pandas import Series,DataFrame
import os

#učitavanje u dataframe
df = pd.read_csv("kontrola.csv", sep = ",")
df = df.drop_duplicates()
df = df.reset_index(drop=True)
indexi = []

#iteriranje kroz dataframe i hvatanje indexa reda u kojem su u istoj tekmi igrači home i away
for index, item in df.iterrows():
    precursor_row = df.iloc[index-1]
    if((int(item['id_igraca']) == int(precursor_row.id_igraca)) & (str(item['tim']) != str(precursor_row.tim)) &
      (str(item['domaci']) == str(precursor_row.domaci)) & (str(item['gosti']) == str(precursor_row.gosti)) &
      (int(item['id_utakmice']) == int(precursor_row.id_utakmice)) & 
       ((str(item['tim']) == 'home') & (str(precursor_row.tim) == 'away') & 
           (int(item['broj_akcija']) < int(precursor_row.broj_akcija)))):
        indexi.append(index)
    elif((int(item['id_igraca']) == int(precursor_row.id_igraca)) & (str(item['tim']) != str(precursor_row.tim)) &
      (str(item['domaci']) == str(precursor_row.domaci)) & (str(item['gosti']) == str(precursor_row.gosti)) &
      (int(item['id_utakmice']) == int(precursor_row.id_utakmice)) & 
       ((str(item['tim']) == 'home') & (str(precursor_row.tim) == 'away') & 
           (int(item['broj_akcija']) > int(precursor_row.broj_akcija)))):
        indexi.append(index)
    elif((int(item['id_igraca']) == int(precursor_row.id_igraca)) & (str(item['tim']) != str(precursor_row.tim)) &
      (str(item['domaci']) == str(precursor_row.domaci)) & (str(item['gosti']) == str(precursor_row.gosti)) &
      (int(item['id_utakmice']) == int(precursor_row.id_utakmice)) & 
       ((str(item['tim']) == 'away') & (str(precursor_row.tim) == 'home') & 
           (int(item['broj_akcija']) < int(precursor_row.broj_akcija)))):
        indexi.append(index)
    elif((int(item['id_igraca']) == int(precursor_row.id_igraca)) & (str(item['tim']) != str(precursor_row.tim)) &
      (str(item['domaci']) == str(precursor_row.domaci)) & (str(item['gosti']) == str(precursor_row.gosti)) &
      (int(item['id_utakmice']) == int(precursor_row.id_utakmice)) & 
       ((str(item['tim']) == 'away') & (str(precursor_row.tim) == 'home') & 
           (int(item['broj_akcija']) < int(precursor_row.broj_akcija)))):
        indexi.append(index)
            
#df.to_csv("kontrola2.csv", sep = ",", index=False)
print(indexi)
```

```python
#kreiranje dataframe sa igračima i klubovima
import numpy as np
import seaborn as sns
import pandas as pd
from pandas import Series,DataFrame
import os

df4 = pd.read_csv("filtrirana_kontrola.csv", sep=",")

#novi dataframe
igraci_klubova = pd.DataFrame(columns=['id_igraca', 'id_kluba'])

for index, item in df4.iterrows():
    if(str(item['tim']) == 'away'):
        igraci_klubova.set_value(index, 'id_igraca', int(item['id_igraca']))
        igraci_klubova.set_value(index, 'id_kluba', int(item['id_gosti']))
    elif(str(item['tim']) == 'home'):
        igraci_klubova.set_value(index, 'id_igraca', int(item['id_igraca']))
        igraci_klubova.set_value(index, 'id_kluba', int(item['id_domaci']))
        
igraci_klubova = igraci_klubova.drop_duplicates(subset='id_igraca')
igraci_klubova = igraci_klubova.reset_index(drop=True)
igraci_klubova.to_csv("igraci_klubova.csv", sep=",", index=False)

#10347 je id Eintrachta
klub = igraci_klubova['id_kluba'] == 10347

ig = igraci_klubova[klub]
ig = ig.reset_index(drop=True)
ig
    
```
