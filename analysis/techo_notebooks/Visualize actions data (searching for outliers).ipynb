```python
import matplotlib
import pandas as pd
import matplotlib.pyplot as plt
from pandas.tools.plotting import scatter_matrix
import warnings
warnings.filterwarnings('ignore')

dfEintrachtUkupno = pd.read_csv("Eintracht_ukupno.csv", sep =',')

dfEintrachtClean =   dfEintrachtUkupno.drop(['actionStartTime', 'actionEndTime', 'goalZone', 'refPositionX',
  'refPositionY', 'startPossTime', 'endPossTime' , 'gameTimeSection', 'awayTeamSide',
  'actionIDNaziv', 'bindedActionIDN', 'actionOptionIDN', 'playerPositionIDN',
  'targetPositionIDN', 'domaciFormacija', 'gostiFormacija', 'gameTime', 'stadium',
  'referee', 'climate', 'luna', 'temperature', 'viewers', 'Pogodaka_domaci',
  'Pogodaka_gosti', 'tocno_dodavanje_class'], axis = 1)

dfEintrachtClean
```

```python
fig1 = plt.figure(1, figsize = (22,22))
ax = fig1.gca()
sm = scatter_matrix(dfEintrachtClean, alpha=0.2, figsize=(15, 15), diagonal='hist', ax=ax)

[s.xaxis.label.set_rotation(45) for s in sm.reshape(-1)]
[s.yaxis.label.set_rotation(0) for s in sm.reshape(-1)]

[s.get_yaxis().set_label_coords(-0.7,0.5) for s in sm.reshape(-1)]
fig1.savefig('ScatterMatrica.png')
plt.show()
```
